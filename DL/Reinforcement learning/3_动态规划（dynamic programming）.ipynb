# 1 悬崖漫步（Cliff Walking）

# 1.1 悬崖漫步环境

```python
import copy


class CliffWalkingEnv:
    """ 悬崖漫步环境"""
    def __init__(self, ncol=12, nrow=4):
        self.ncol = ncol  # 定义网格世界的列
        self.nrow = nrow  # 定义网格世界的行
        # 转移矩阵P[state][action] = [(p, next_state, reward, done)]包含下一个状态和奖励
        self.P = self.createP() # (48,4,1,4)

    def createP(self):
        # 初始化
        P = [[[] for j in range(4)] for i in range(self.nrow * self.ncol)] # shape=(nrow*ncol, 4, 1/0),0因为是空表格
        # 4种动作, change[0]:上,change[1]:下, change[2]:左, change[3]:右。坐标系原点(0,0)
        # 定义在左上角
        change = [[0, -1], [0, 1], [-1, 0], [1, 0]]
        for i in range(self.nrow):
            for j in range(self.ncol):
                for a in range(4):
                    # 位置在悬崖或者目标状态,因为无法继续交互,任何动作奖励都为0
                    if i == self.nrow - 1 and j > 0:
                        P[i * self.ncol + j][a] = [(1, i * self.ncol + j, 0,
                                                    True)]
                        continue
                    # 其他位置
                    next_x = min(self.ncol - 1, max(0, j + change[a][0]))
                    next_y = min(self.nrow - 1, max(0, i + change[a][1]))
                    next_state = next_y * self.ncol + next_x
                    reward = -1
                    done = False
                    # 下一个位置在悬崖或者终点
                    if next_y == self.nrow - 1 and next_x > 0:
                        done = True
                        if next_x != self.ncol - 1:  # 下一个位置在悬崖
                            reward = -100
                    P[i * self.ncol + j][a] = [(1, next_state, reward, done)]
        return P
```

# 1.2 策略迭代

```python
class PolicyIteration:
    """ 策略迭代算法 """
    def __init__(self, env, theta, gamma):
        self.env = env
        # 初始化每个状态的状态价值为0
        self.v = [0] * self.env.ncol * self.env.nrow  
        # 将在每个方格(状态)选取不同动作的策略初始化为均匀随机策略
        self.pi = [[0.25, 0.25, 0.25, 0.25] for i in range(self.env.ncol * self.env.nrow)]  # (48, 4)
        self.theta = theta  # 策略评估收敛阈值
        self.gamma = gamma  # 折扣因子

    def policy_evaluation(self):  
        '''
        策略评估
        '''
        cnt = 1  # 计数器
        while 1:
            max_diff = 0
            new_v = [0] * self.env.ncol * self.env.nrow
            # 循环每个状态，把某个策略下，每一个状态的状态价值函数都计算一下
            for s in range(self.env.ncol * self.env.nrow):   
                # 利用贝尔曼期望方程计算每个状态s下的状态价值函数，并不断更新self.v
                qsa_list = []  
                # 循环当前状态的每个动作，计算并更新当前状态的状态价值
                for a in range(4):
                    # 在状态s和动作a下的返回值仅为一个元组-状态转移概率，下个状态，即时奖励，是否结束.
                    for res in self.env.P[s][a]: 
                        p, next_state, r, done = res
                        # 计算当前状态下采取当前动作的动作价值函数Q(s,a)，
                        # 本章环境比较特殊,奖励和下一个状态有关,所以需要和状态转移概率相乘
                        qsa = p * (r + self.gamma * self.v[next_state] * (1 - done))
                    # Q(s,a)分别乘以π(a|s)，在后面代码中对乘法的结果进行累加得到该状态的状态价值函数。
                    qsa_list.append(self.pi[s][a] * qsa) # 计算一个状态下，所有可能的确定动作的动作价值函数
                # 在当前策略下，当前状态的状态价值函数。对Q(s,a)乘π(a|s)的结果进行累加得到该状态的状态价值函数。
                # 状态价值函数和动作价值函数之间的关系
                new_v[s] = sum(qsa_list)
                # 当某个状态的状态价值函数达到收敛条件则作为退出循环。
                max_diff = max(max_diff, abs(new_v[s] - self.v[s])) 
            # 不断更新当前策略的状态价值函数
            # 若满足收敛条件,则退出策略评估迭代，并返回该策略对每个状态的状态价值函数
            self.v = new_v
            # 满足收敛条件,退出评估迭代
            if max_diff < self.theta: break 
            cnt += 1
        print("策略评估进行%d轮后完成" % cnt)

    def policy_improvement(self):  # 策略提升
        # 对每个状态循环一遍
        for s in range(self.env.nrow * self.env.ncol):
            qsa_list = []
            for a in range(4):
                for res in self.env.P[s][a]:
                    p, next_state, r, done = res
                    # 计算当前状态下采取当前动作的动作价值函数Q(s,a)，
                    # 本章环境比较特殊,奖励和下一个状态有关,所以需要和状态转移概率相乘
                    qsa = p * (r + self.gamma * self.v[next_state] * (1 - done))
                qsa_list.append(qsa)
            # 获得当前状态下所有动作中最大的动作价值
            maxq = max(qsa_list)
            # 当前状态中，有几个动作得到相同的最大Q值
            cntq = qsa_list.count(maxq)  
            # 让这些动作均分概率
            self.pi[s] = [1 / cntq if q == maxq else 0 for q in qsa_list]
        print("策略提升完成")
        return self.pi

    def policy_iteration(self):  
        '''
        策略迭代
        '''
        while 1:
            # 策略评估，不断对当前策略下每个状态的状态价值迭代更新，
            # 当某个状态的的状态价值函数达到收敛条件则作为退出循环。
            self.policy_evaluation()
            # 将列表进行深拷贝,方便接下来进行比较
            old_pi = copy.deepcopy(self.pi)  
            # 策略提升
            # 对每个状态循环一遍，直接对每个状态下不同动作的选取使用贪心法则，
            # 即使得每个状态下动作价值最大的动作的选取概率为1或$\frac{1}{最大Q的个数}$，其他都为0
            new_pi = self.policy_improvement()
            if old_pi == new_pi: break
                
                
                
def print_agent(agent, action_meaning, disaster=[], end=[]):
    print("每个状态的状态价值：")  
    for i in range(agent.env.nrow):
        for j in range(agent.env.ncol):
            # 为了输出美观,保持输出6个字符
            print('%6.6s' % ('%.3f' % agent.v[i * agent.env.ncol + j]), end=' ')
        print()

    print("策略(某个确定状态有可能采取到的所有动作)：")
    for i in range(agent.env.nrow):
        for j in range(agent.env.ncol):
            # 一些特殊的状态,例如悬崖漫步中的悬崖
            if (i * agent.env.ncol + j) in disaster:
                print('****', end=' ')
            elif (i * agent.env.ncol + j) in end:  # 目标状态
                print('EEEE', end=' ')
            else:
                a = agent.pi[i * agent.env.ncol + j]
                pi_str = ''
                for k in range(len(action_meaning)):
                    pi_str += action_meaning[k] if a[k] > 0 else 'o'  # 输出策略，即在某个状态有可能采取到的动作
                print(pi_str, end=' ')
        print()


env = CliffWalkingEnv()
action_meaning = ['^', 'V', '<', '>']
theta = 0.001
gamma = 0.9
agent = PolicyIteration(env, theta, gamma)
agent.policy_iteration()
print_agent(agent, action_meaning, list(range(37, 47)), [47])
```

# 1.3 价值迭代

```python
class ValueIteration:
    """ 价值迭代算法 """
    def __init__(self, env, theta, gamma):
        self.env = env
        self.v = [0] * self.env.ncol * self.env.nrow  # 初始化价值为0。
        self.theta = theta  # 价值收敛阈值
        self.gamma = gamma
        # 价值迭代结束后得到的策略
        self.pi = [None for i in range(self.env.ncol * self.env.nrow)] # (48,4)

    def value_iteration(self):
        cnt = 0
        while 1:
            max_diff = 0
            new_v = [0] * self.env.ncol * self.env.nrow
            # 不断循环每个状态，对状态价值进行更新，直到某个状态的状态价值达到收敛条件
            for s in range(self.env.ncol * self.env.nrow):
                # 存储所有状态s的所有Q(s,a)价值
                qsa_list = [] 
                # 循环当前状态的每个动作，计算并更新当前状态的状态价值
                for a in range(4):
                    for res in self.env.P[s][a]:
                        p, next_state, r, done = res
                        qsa = p * (r + self.gamma * self.v[next_state] * (1 - done))
                    # 下两行代码是价值迭代和策略迭代的主要区别
                    qsa_list.append(qsa)  
                # 选取当前状态中使得动作价值最大的动作的动作价值函数去更新当前状态价值，即选取最优状态价值
                # 该选取规则是贝尔曼最优方程，即每个状态的状态价值函数只进行一轮更新，然后直接根据当前状态更新后的价值进行策略提升
                new_v[s] = max(qsa_list)
                max_diff = max(max_diff, abs(new_v[s] - self.v[s]))
            self.v = new_v
            # 满足收敛条件,退出价值函数迭代
            if max_diff < self.theta: break  
            cnt += 1
        print("价值迭代一共进行%d轮" % cnt)
        self.get_policy()

    def get_policy(self):  
        '''
        价值迭代结束后，根据价值函数返回导出一个贪婪策略
        '''
        # 通过对每个状态不断的循环，获得每个状态下选择某个动作的最优策略
        for s in range(self.env.nrow * self.env.ncol):
            qsa_list = []
            for a in range(4):
                qsa = 0
                for res in self.env.P[s][a]:
                    p, next_state, r, done = res
                    qsa += p * (r + self.gamma * self.v[next_state] * (1 - done))
                qsa_list.append(qsa)
            maxq = max(qsa_list)
            # 计算有几个动作得到了最大的Q值
            cntq = qsa_list.count(maxq)  
            # 让这些动作均分概率
            self.pi[s] = [1 / cntq if q == maxq else 0 for q in qsa_list]


env = CliffWalkingEnv()
action_meaning = ['^', 'v', '<', '>']
theta = 0.001
gamma = 0.9
agent = ValueIteration(env, theta, gamma)
agent.value_iteration()
print_agent(agent, action_meaning, list(range(37, 47)), [47])
```

# 2.1 环境

```python
import gym
env = gym.make("FrozenLake-v1")  # 创建环境
env = env.unwrapped  # 解封装才能访问状态转移矩阵P
env.render()  # 环境渲染,通常是弹窗显示或打印出可视化的环境

holes = set()
ends = set()
for s in env.P:
    for a in env.P[s]:
        for s_ in env.P[s][a]:
            if s_[2] == 1.0:  # 获得奖励为1,代表是目标
                ends.add(s_[1])
            if s_[3] == True:
                holes.add(s_[1])
holes = holes - ends
print("冰洞的索引:", holes)
print("目标的索引:", ends)

for a in env.P[14]:  # 查看目标左边一格的状态转移信息
    print(env.P[14][a])
```
