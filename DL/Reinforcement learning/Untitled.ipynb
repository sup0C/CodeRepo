```python
import torch
t=torch.tensor([1,2,3,4,5],dtype=torch.float32)
mean=0.0; std=0.01
torch.nn.init.normal_(t, mean=mean, std=std)
cond = (t < mean - 2 * std) | (t > mean + 2 * std)
a=not torch.sum(cond)
print(cond,mean - 2 * std,mean + 2 * std,t,t > mean + 2 * std,a)
```
